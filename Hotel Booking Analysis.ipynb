# Data Wrangling


```python
#creating a duplicate of the original dataset 
hb_df_new = hb_df.copy()
```

```python
hb_df_new.columns
```

```python
#replacing null values in children column with 0 assuming that family had 0 children and similarly for company adn agent columns
hb_df_new['children'].fillna(0, inplace = True)
hb_df_new['company'].fillna(0, inplace = True)
hb_df_new['agent'].fillna(0, inplace = True)

#replacing null values in country columns as Others

hb_df_new['country'].fillna('Others', inplace = True)
```

```python
hb_df_new.isnull().sum()
```

```python
#dropping the company column as it contains a lot of null values compares with others
hb_df_new.drop(['company'], axis = 1, inplace = True)
```

```python
#dropping rows with 0 adults, children and babies
no_guest = hb_df_new[hb_df_new['adults'] + hb_df_new['babies'] + hb_df_new['children'] == 0]
hb_df_new.drop(no_guest.index, inplace = True)
```

```python
# adding some new columns to make our data analysis ready
hb_df_new['total_people'] = hb_df_new['adults'] + hb_df_new['babies'] + hb_df_new['children']       # creating total people column by adding all the people in that booking

hb_df_new['total_stay'] = hb_df_new['stays_in_weekend_nights'] + hb_df_new['stays_in_week_nights']    # creating a column to check tota
```

# Data Visualisation

# Which type of hotel is most preferred by the guests?



```python
#storing hotel names in a variable
hotel_name = hb_df_new['hotel'].unique()

#checking the number of unique booking in each hotel type
unique_booking = hb_df_new.hotel.value_counts().sort_values(ascending = True)

#Creating a pie chart using plotly.express
fig1 = px.pie(names = hotel_name, values = unique_booking, hole = 0.5, color = hotel_name,
             color_discrete_map = {'Resort Hotel' : 'teal', 'City Hotel': 'green'})

#Giving it a title and updating the text info
fig1.update_traces(textinfo = 'percent + value')
fig1.update_layout(title_text = 'Hotel Booking Percentage', title_x = 0.5)

#setting the Legend at Centre
fig1.update_layout(legend = dict(
                            orientation = 'h', yanchor = 'bottom',
                            xanchor = 'center', x = 0.5))

fig1.show()
```

```python
# Count Hotel
hotel_count = hb_df_new.hotel.value_counts()

# Plotting Values in a simple pie chart
hotel_count.plot.pie(figsize=(9,8), autopct='%1.2f%%', shadow=True, fontsize=15,startangle=50)
# Setting the title
plt.title('Hotel Booking Percentage')
plt.axis('equal')
plt.show()
```

chart 5 
Top 5 agents who has most number of booking

```python
# Creating a dataset by grouping by agent column and it's count
agents = hb_df_new.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count':'Booking Count'}
                                                                                ).sort_values(by = 'Booking Count', ascending = False)

# Extracting top 5 agents by booking count
top_5 = agents[:5]

# Explosion
explode = (0.02,0.02,0.02,0.02,0.02)

# Colors
colors = ( "orange", "cyan", "brown", "indigo", "beige")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Plotting the values
fig, ax = plt.subplots(figsize =(15, 7))
wedges, texts, autotexts = ax.pie(top_5['Booking Count'],
                                  autopct = lambda pct: func(pct, top_5['Booking Count']),
                                  explode = explode,
                                  shadow = False,
                                  colors = colors,
                                  startangle = 50,
                                  wedgeprops = wp)

# Adding legend
ax.legend(wedges, top_5['agent'],
          title ="agents",
          loc ="upper left",
          bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Top 5 agents in terms of booking", fontsize = 17)

# Show chart
plt.axis('equal')
plt.show()
     
```

What is the adr across different months?

```python
# Grouping arrival_month and hotel on mean of adr
bookings_months=hb_df_new.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

# Creating a month list to order the months in ascending
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Creating a dataset of months, hotel and their adr
bookings_months['arrival_date_month']=pd.Categorical(bookings_months['arrival_date_month'],categories=months,ordered=True)

# Sorting the months
bookings_months=bookings_months.sort_values('arrival_date_month')
bookings_months
```

```python
# Setting the chart size
plt.figure(figsize=(15,5))

# Plotting the values in a line chart
sns.lineplot(x=bookings_months['arrival_date_month'],y=bookings_months['adr'],hue=bookings_months['hotel'])

# Setting the labels and title
plt.title('ADR across each month', fontsize=20)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('ADR', fontsize=12)

# Show chart
plt.show()
```

```python
# Grouping dist_channel and hotels on their adr
dist_channel_adr = hb_df_new.groupby(['distribution_channel','hotel'])['adr'].mean().reset_index()

# Setting the figure size
plt.figure(figsize=(15,5))

# Creating a horizontal bar chart
sns.barplot(x='adr', y='distribution_channel', data=dist_channel_adr, hue='hotel')

# Setting the title
plt.title('ADR across each distribution channel', fontsize=20)

# Show chart
plt.show()
```
